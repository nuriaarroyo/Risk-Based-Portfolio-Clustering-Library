```python
from portafolios.core.portafolio import Portfolio
from portafolios.data.local_loader import local_loader
from portafolios.constructores.naive import Naive
from portafolios.constructores.markowitz import Markowitz
```

```python


csv_path = "data_clean_stock_data.csv"

# loader que ignora tickers/start/end y lee TU csv
loader_from_csv = lambda *args, **kwargs: local_loader(*args, **kwargs, path=csv_path)

p = (Portfolio(
        tickers=None,
        start="2019-01-01",
        end="2019-12-31",
        loader=local_loader,
        loader_kwargs={"path": csv_path, "prefer_adj_close": True},
        freq="B",
     )
     .preparar_datos()
                  # ← capa de métricas por activo
)
```

```python


mk = Markowitz(rf_per_period=0.0)

# 1) Versión "teoría" (default) usando log(1 + r_simple)
p.construir(mk)                 # ret_kind="log" por default

# 2) Versión usando retornos simples tal cual
p.construir(mk, ret_kind="simple")

# y luego:
p.kpis_basicos(ann_factor=252, rf_per_period=0.0)
p.pastel()
p.barras()
p.bubbleplot()
```

```python
from pathlib import Path
import webbrowser

# Carpeta donde están todos tus HTML
BASE_PATH = Path(r"c:\Users\narro\Desktop\semestre\honores_actuaria\plots")

def ver_html(nombre_archivo: str):
    """
    Abre en el navegador un archivo HTML ubicado en la carpeta BASE_PATH.
    Parámetros:
        nombre_archivo (str): nombre del archivo HTML, ej: 'grafica.html'
    """
    ruta = BASE_PATH / nombre_archivo
    if not ruta.exists():
        raise FileNotFoundError(f"No se encontró el archivo: {ruta}")

    url = ruta.resolve().as_uri()
    webbrowser.open(url)
    print(f"Abriendo: {url}")
```
