```python
from pathlib import Path
import webbrowser

# Carpeta donde están todos tus HTML
BASE_PATH = Path(r"c:\Users\narro\Desktop\semestre\honores_actuaria\plots")

def ver_html(nombre_archivo: str):
    """
    Abre en el navegador un archivo HTML ubicado en la carpeta BASE_PATH.
    Parámetros:
        nombre_archivo (str): nombre del archivo HTML, ej: 'grafica.html'
    """
    ruta = BASE_PATH / nombre_archivo
    if not ruta.exists():
        raise FileNotFoundError(f"No se encontró el archivo: {ruta}")

    url = ruta.resolve().as_uri()
    webbrowser.open(url)
    print(f"Abriendo: {url}")
```
