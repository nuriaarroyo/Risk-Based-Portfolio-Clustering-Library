# Portfolio loader examples

This notebook shows the same `Portfolio` structure using:
- the local CSV loader
- the Yahoo Finance loader

In both cases we use the same callable: `loader=load_prices`.

```python
from portafolios import Portfolio, load_prices

from pathlib import Path

PROJECT_ROOT = Path.cwd().resolve()
while not (PROJECT_ROOT / "portafolios").exists() and PROJECT_ROOT != PROJECT_ROOT.parent:
    PROJECT_ROOT = PROJECT_ROOT.parent

CSV_PATH = PROJECT_ROOT / "data" / "processed" / "data_clean_stock_data.csv"
YF_SNAPSHOT_PATH = PROJECT_ROOT / "data" / "yf_snapshot.csv"

print("Project root:", PROJECT_ROOT)
print("Processed CSV exists:", CSV_PATH.exists())
print("Yahoo snapshot exists:", YF_SNAPSHOT_PATH.exists())
```

## Local loader example

```python
portfolio_local = Portfolio(
    tickers=["AAPL", "MSFT", "AMZN"],
    start="2020-01-01",
    end="2020-06-30",
    loader=load_prices,
    loader_kwargs={
        "source": "local",
        "path": CSV_PATH,
        "prefer_adj_close": True,
    },
)

portfolio_local.preparar_datos()
portfolio_local.prices.head()
```

```python
portfolio_local.prices.shape
```

## Yahoo Finance loader example

```python
portfolio_yf = Portfolio(
    tickers=["AAPL", "MSFT", "AMZN"],
    start="2024-01-01",
    end="2024-03-31",
    loader=load_prices,
    loader_kwargs={
        "source": "yfinance",
        "prefer_adj_close": True,
        "batch_size": 100,
        "progress": False,
        "save_path": YF_SNAPSHOT_PATH,
        "fallback_to_saved_data": True,
    },
)

portfolio_yf.preparar_datos()
portfolio_yf.prices.head()
```

```python
portfolio_yf.prices.shape
```

## Quick comparison

Only `loader_kwargs` changes. The `Portfolio` structure is the same.

```python
print("Local columns:", portfolio_local.prices.columns.tolist())
print("Yahoo columns:", portfolio_yf.prices.columns.tolist())
```
