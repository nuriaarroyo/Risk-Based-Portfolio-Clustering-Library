This notebook gives a visual scheme of the current `portafolios` library.

- which pieces are core and which pieces are more legacy / auxiliary


## 1. High-level architecture

This figure answers:
- where data enters
- where constructions are built
- where backtests / Monte Carlo are attached

## 3. Practical reading of the architecture

The cleanest current path is:

1. `CSVLoader` or `YFinanceLoader` produces standardized market data.
2. `PortfolioUniverse` stores shared prices, returns, metadata, and output paths.
3. Constructors build fixed weights and save them as `ConstructionResult` objects.
4. `Backtester` and `MonteCarloEngine` evaluate those fixed weights later.
5. `PortfolioVisualizer` reads the stored results and makes Plotly figures.
6. Exports are organized in `outputs/runs/<universe_name>/...`

The main remaining conceptual rough spot is that `core/portafolio.py` still has several responsibilities together:
- main experiment container
- storage/export coordinator
- some legacy plotting passthroughs

So visually the library already has the right scope, but the central class is still carrying a lot of the orchestration.
